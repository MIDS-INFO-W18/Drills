We won't need the Indicator_Name and Ind_Code columns. So remove 'em.

```python
import pandas as pd
import numpy as np
df2 = dataFile.drop(['Indicator_Name','Ind_Code'], axis=1)
df2 = df2.set_index('Country_Name')
df2.sort_index(inplace=True)
df2
```

The data range from 1995 to 2016.  We want to create two new columns: one for 1995-1999 and one from 2000-2016.  Let's call them 20thCent and 21stCent.

```python
oldCent = ['1995','1995','1996','1997','1998','1999']
df2[df2.columns[6:22]]
df2['19thCent'] = df2[oldCent].sum(axis=1)
df2['20thCent'] = df2[df2.columns[6:22]].sum(axis=1)
df2
```
