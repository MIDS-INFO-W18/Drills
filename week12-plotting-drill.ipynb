### Line Plot
Like most of the plots, we require data about the axes, labels, a title, and data!  In this example, notice that we create an object to hold the plot (called ax), then assign the data (best fit, line width (lw), the color (expressed in an rgb hex triplet), and the alpha channel (or transparency, between 0.01 and 1).

```python
import matplotlib.pyplot as plt
import numpy as np

# define a function for our line graph with the necessary parameters
# after the function, add your data and text info ... experiment with 
# colors, line width, and transparency
def lineplot(x_data, y_data, x_label="", y_label="", title=""):
    # Create the plot object
    _, ax = plt.subplots()

    # Plot the best fit line, set the linewidth (lw), color and
    # transparency (alpha) of the line
    ax.plot(x_data, y_data, lw = 2, color = '#539caf', alpha = 1)

    # Label the axes and provide a title
    ax.set_title(title)
    ax.set_xlabel(x_label)
    ax.set_ylabel(y_label)

# You provide the following:
# Title of the plot
# x data
# y data
# the labels for the x and y axes
# title?

# now run and explore:
# 1)  run by passing parameters
# lineplot([.31,2,3],[3,4,5],"fish","cat","Meow")

# 2) run by integrating numpy and data in a numpy multi array
# SOLUTION:
#data = [[2,.2,12,42,32],[8,6,5,6,1],[6,6,5,9,14]]
#data = np.array(data)
#plt.plot(np.arange(1,6), data.transpose())
#plt.show()

#3) run the same line graph, reading data in from a .json file.
# SOLUTION: tba
```

### Scatter Plots
These are really useful and can be quite artistic.  In this exercise we'll practice the creation of a scatter plot - and you'll see the commonalities between the syntax of the line and the scatter plot scripts. We may need to plot a lot or very little data ... you may want to try a log scale for the y-axis ...  Please note that when plotting data, use the size of the icon to represent the volume or number of items in that group - for example, a scatter plot may have more than one color to represent subgroups, but the size of each circle represents the number of discrete items in that group.  A small circle might have a singleton; a large may hold 100s.

```python
import matplotlib.pyplot as plt
import numpy as np

def scatterplot(x_data, y_data, x_label="", y_label="", title="", color = "r", yscale_log=False):

    # Create the plot object - ax for "axis data"
    _, ax = plt.subplots()

    # Plot the data, set the size (s), color and transparency (alpha)
    # of the points
    ax.scatter(x_data, y_data, s = 150, color = color, alpha = 0.75)

    if yscale_log == True:
        ax.set_yscale('log')  # perhaps a log scale

    # Label the axes and provide a title
    ax.set_title(title)
    ax.set_xlabel(x_label)
    ax.set_ylabel(y_label)

# for a first test supply your own data in detail ...
# SOLUTION:
#x_data = [30,40,50,60,70,80,90,100,110]
#y_data = [5,5,2,3,35,52,87,92,76]
#x_label= "temperature"
#y_label = "murder_rate"
#title="Is there a relationship between temp and behavior?"

# for more fun, let's make the chart more attractive and plot 
# more than one data set...
# create a figure object ("fig") to which we can add the plot data
# we'll need a reference to an Axes object to draw on the same subplot.
# SOLUTION:
#x1 = [10,3,4,2,4]
#x2 = [5,3,2,4,1]
#y1 = [3,1,1,4,3]
#y2 = [8,7,6,7,8]
#fig = plt.figure()

#ax1 = fig.add_subplot(111)
#ax1.scatter(x1,x2, s=375, c='b', alpha=.5, marker="s", label='round')
#ax1.scatter(y1,y2, s=370, c='r', alpha=.25, marker="o", label='square')
# add a legend/key
#plt.legend(loc='upper right')
# show the plot
#plt.show()
```

### Histograms
Histograms are useful for seeing distributions.  Control the number of bars in the histogram by setting the number of "bins" to use - matplotlib's n_bins.  The cumulative parameters lets you select between probability density function (pdf) or cumulative density function (cdf). 
As above, we may have more than one data set or group that overlay each other.  Consequently, we adjust the color transparency.

```python
import matplotlib.pyplot as plt
import numpy as np

# Overlay 2 histograms to compare them
def overlaid_histogram(data1, data2, n_bins = 0, data1_name="", 
    data1_color="#539caf", data2_name="", data2_color="#7663b0", 
    x_label="", y_label="", title=""):
    
    # Set the bounds for the bins so that the two distributions are fairly compared
    max_nbins = 10
    data_range = [min(min(data1), min(data2)), max(max(data1), max(data2))]
    binwidth = (data_range[1] - data_range[0]) / max_nbins


    if n_bins == 0:
    	bins = np.arange(data_range[0], data_range[1] + binwidth, binwidth)
    else: 
    	bins = n_bins

    # Create the plot
    _, ax = plt.subplots()
    ax.hist(data1, bins = bins, color = data1_color, alpha = 1, label = data1_name)
    ax.hist(data2, bins = bins, color = data2_color, alpha = 0.75, label = data2_name)
    ax.set_ylabel(y_label)
    ax.set_xlabel(x_label)
    ax.set_title(title)
    ax.legend(loc = 'best')

# adjust the data in these vars for practice - hard-coded, read from file, etc.
data1 = [.25, .25, .25, .3, .35, 1, 1.25, 1.6, 1.9, 1, .9, .5]
data2 = [.15, .20, .3, .31, .35, 1.5, 1.75, 1.7, 1.3, 1.2, .7, .2]
n_bins = 12
data1_name="Group 1"
data1_color="#539caf"
data2_name="Group 2" 
data2_color="#7663b0"
x_label="No of participants",
y_label="Average Scores", 
title="Independent Means (2 groups)"
         
overlaid_histogram(data1, data2, n_bins, data1_name, data1_color, data2_name, data2_color, 
                   x_label, y_label, title)
```

### Heatmap 


```python
import pandas as pd
from pandas import Series
from pandas import DataFrame
from pandas import TimeGrouper
from matplotlib import pyplot
series = pd.read_csv('~/Documents/UCB-DataSci/fakedata.csv', header=0)
groups = series.groupby(TimeGrouper('A'))
years = DataFrame()
for name, group in groups:
	years[name.year] = group.values
years = years.T
pyplot.matshow(years, interpolation=None, aspect='auto')
pyplot.show()

# https://machinelearningmastery.com/time-series-data-visualization-with-python/
```
